```python
# 20. Classify Movies Based on Ratings [Good,Better and Best]
df.columns
```

```python
def Rating(rating):
    if rating >= 7.0:
        return "Excellent"
    elif rating >= 6.0:
        return "Good"
    else:
        return "Average"
```

```python
df["Rating_Category"] = df['Rating'].apply(Rating)
```
