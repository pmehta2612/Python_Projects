```python
# 8. Draw Correlation Matrix 
plt.figure(figsize = (17,6))
sns.heatmap(df.corr(), annot = True)
plt.title(label = "Correlation Matrix")
plt.show()
```

```python
# 9. How Many People Have Heart Disease, And How Many Don't Have Heart Disease In This Dataset?
df['target'].value_counts()
```

```python
# Chart Design Theme
print(plt.style.available)
plt.style.use("seaborn-v0_8")
```

```python
# Viewing using visualization
sns.countplot(data = df, x = 'target')
plt.xticks([0,1],['No Disease','Heart Disease'])
plt.title(label = "Heart Disease Patients v/s No Heart Disease Patients")
plt.show()
```

```python
# 10. Find Count of  Male & Female in this Dataset
df['sex'].value_counts()
```

```python
# Viewing using visualization
sns.countplot(data = df, x = 'sex')
plt.xticks([0,1],['Female','Male'])
plt.title(label = "Male v/s Female Patients")
plt.xlabel(xlabel = "Gender")
plt.show()
```

```python
# 11. Find Gender Distribution According to The Target Variable
df.columns
```

```python
sns.countplot(data = df, x = 'sex', hue = df['target'].astype(str))
plt.xticks([0,1],['Female','Male'])
plt.legend(labels = ['No-Disease','Heart-Disease'])
plt.title(label = "Gender Distribution")
plt.show()
```

```python
# 12. Check age distribution in the dataset
plt.title(label = "Age Distribution")
sns.distplot(df['age'], bins = 20, color = "Blue")
plt.show()
```

```python
# 13. Check chest pain type
plt.figure(figsize = (10,4))
sns.countplot(data = df, x = 'cp')
plt.xticks([0,1,2,3],["Typical Angina","Atypical Angina","Non-Anginal Pain","Aymptomatic"])
plt.title(label = "Chest Pain Type Distribution")
plt.xlabel(xlabel = "Chest Pain Type")
plt.show()
```

```python
# 14. Show the chest pain distribution as per target variable
plt.figure(figsize = (10,5))
sns.countplot(data = df, x = 'cp', hue = df['target'].astype(str))
plt.xticks([0,1,2,3],["Typical Angina","Atypical Angina","Non-Anginal Pain","Aymptomatic"])
plt.legend(labels = ["No-Disease","Heart Disease"])
plt.title(label = "Chest Pain Type Distribution on the Basis of Disease")
plt.xlabel(xlabel = "Chest Pain Type")
plt.show()
```

```python
# 15. Show fasting blood sugar distribution according to target variable
plt.figure(figsize = (10,5))
sns.countplot(data = df, x = 'fbs', hue = df['target'].astype(str))
plt.xticks([0,1],["No","Yes"])
plt.title(label = "Fasting Blood Sugar Distribution")
plt.xlabel(xlabel = "Fasting Blood Sugar")
plt.legend(labels = ["No-Disease","Heart Disease"])
plt.show()
```

```python
# 16. Check resting blood pressure distribution
df['trestbps'].hist(grid = True)
plt.title(label = "Resting Blood Pressure Distribution")
plt.show()
```

```python
# 17. Compare resting blood pressure as per sex column
#plt.style.use("ggplot")
g = sns.FacetGrid(df,hue = 'sex',aspect=4)
g.map(sns.kdeplot,'trestbps',fill = True)
plt.legend(labels = ['Male','Female'])
plt.title(label = "Comparing Resting Blood Pressure by Gender")
plt.show()
```

```python
# 18. Show distribution of Serum Cholesterol
df['chol'].hist()
plt.title(label = "Serum Cholesterol Distribution")
```

```python
# 19. Plot Continuous Variables
cate_value =[]
cont_value =[]

for column in df.columns:
    if df[column].nunique() <= 10:
        cate_value.append(column)
    else:
        cont_value.append(column)
```

```python
print("Categorical Variables are: ",cate_value)
print("Continuous Variables are: ",cont_value)
```

```python
df.hist(cont_value, figsize = (10,10))
plt.tight_layout()
plt.show()
```
